# Question 1 - Bootstrap

```julia
import Pkg
Pkg.add("CSV")
Pkg.add("DataFrames")
Pkg.add("Random")
Pkg.add("CategoricalArrays")
Pkg.add("GLM")

using CSV
using DataFrames
using Random
using CategoricalArrays
using GLM
```

```julia
#data = CSV.read("penn_jae.csv", DataFrame)
data = CSV.read("../../data/penn_jae.csv", DataFrame)
data.log_inuidur1 = log.(data.inuidur1)

#Creating T4
data = filter(row -> row.tg in (0, 4), data)
data.T4 = ifelse.(data.tg .== 4, 1, 0)
#dep is categorial
data.dep = categorical(data.dep)
```

### (1) Function to get coefficients 

```julia
#-------------------------------------#
#---Function to get coefficients------#
#-------------------------------------#
function get_coef(data, exp, numlist)

    results = zeros(1, size(numlist)[1])
    model = lm(exp, data)

    for i in 1:size(numlist)[1]
        results[i] = coef(model)[numlist[i]]
    end

    return results

end
#given the position of coefficients in numlist, returns a row vector with each coefficient from a regression. 
```

### (2) Bootstrapping

```julia
#----------------------#
#---Bootstrapping------#
#----------------------#
Random.seed!(1234)
nboot = 10000
point_estimates = zeros(nboot,3) 

for k in 1:nboot

    index = rand(1:size(data)[1], size(data)[1])
    df = data[index, :]
    exp = @formula(log_inuidur1 ~ T4 + (female+black+othrace+dep+q2+q3+q4+q5+q6+agelt35+agegt54+durable+lusd+husd))
    result = get_coef(df, exp, [2,3,4])

    #Appending estimates
    for i in 1:3
        point_estimates[k,i] = result[i]  #10000x3 matrix. Each row contains the desired coefficients from one bootstrap operation
    end

end
```

### (3) Statistics

```julia
#-------------------#
#---Statistics------#
#-------------------#
using Statistics

coef_T4 = point_estimates[:,1]
coef_female = point_estimates[:,2]
coef_black = point_estimates[:,3]

mean_T4 = mean(coef_T4)
mean_female = mean(coef_female)
mean_black = mean(coef_black)

sd_T4 = std(coef_T4, corrected=true)
sd_black = std(coef_black, corrected=true)
sd_female = std(coef_female, corrected=true)
```

### (4) Results

```julia
#--------------#
#---Table------#
#--------------#
using NamedArrays

table = NamedArray(zeros(3, 3))
table[1,2] = mean_T4
table[1,3] = sd_T4
table[2,2] = mean_female
table[2,3] = sd_female
table[3,2] = mean_black
table[3,3] = sd_black

T = DataFrame(table, [ :"Variable", :"Coefficient (boostrap)", :"Standar error (boostrap)"]) 
T[!,:Variable] = string.(T[!,:Variable]) 

T[1,1] = "T4"
T[2,1] = "Female"
T[3,1] = "Black"

println(T)
```
