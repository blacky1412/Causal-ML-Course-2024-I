```julia
using CSV, DataFrames

#---Loading Data------#
#---------------------#
q3 = CSV.File("../../data/wage2015_subsample_inference.csv") |> DataFrame

n = size(q3, 1)
nvar = size(q3, 2)
println(nvar)
q3.id = 1:n  # id variable
println(n)
println(nvar)
```
